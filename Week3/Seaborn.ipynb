# Different visualizations on Seaborn

**Seaborn** is a Python data visualization library built on top of **matplotlib**.  
It provides a high-level interface for creating attractive and informative statistical graphics.  
It works well with pandas DataFrames and makes complex plots easier with less code.

### Main Uses:
- Creating beautiful and informative statistical graphics
- Visualizing distributions, trends, and relationships in data
- Enhancing Matplotlib plots with better default styles and color palettes

---

## 1. Line Plot

**Use Case:**  
Shows how a value changes over a sequence (e.g., time, size, index).  
Great for tracking trends or progress.

---

## 2. Bar Plot

**Use Case:**  
Compares values between different categories.  
Useful to analyze average or total values per group.

---

## 3. Count Plot

**Use Case:**  
Shows how many times each category appears.  
Ideal for visualizing the frequency of categorical data.

---

## 4. Scatter Plot

**Use Case:**  
Displays the relationship between two numeric variables.  
Helps detect patterns or correlations.

---

## 5. Box Plot

**Use Case:**  
Summarizes the distribution of data and highlights outliers.  
Very useful for comparing groups.

---

## 6. Violin Plot

**Use Case:**  
Shows the distribution of the data along with its probability density.  
Combines box plot and KDE.

---

## 7. Swarm Plot

**Use Case:**  
Displays all data points without overlap.  
Best for small datasets where you want to see every point.

---

## 8. Heatmap

**Use Case:**  
Visualizes data in matrix format using colors.  
Commonly used to show correlation between variables.

---

## 9. Histogram (Displot)

**Use Case:**  
Shows how data is distributed over intervals.  
Helps understand frequency distribution of a single variable.

---

## 10. Pair Plot

**Use Case:**  
Shows relationships between all numeric variables in a dataset.  
Useful for quick EDA (exploratory data analysis).

---


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("tips")  # Dataset about restaurant bills and tips
```

```python
sns.lineplot(x="size", y="total_bill", data=df)
plt.title("Line Plot - Total Bill by Group Size")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Bar Plot - Total Bill by Day")
plt.show()
```

```python
sns.countplot(x="day", data=df)
plt.title("Count Plot - Number of Customers per Day")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Box Plot - Total Bill by Day")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
plt.title("Violin Plot - Total Bill by Day")
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=df)
plt.title("Swarm Plot - Total Bill by Day")
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

```python
sns.displot(df["total_bill"], kde=False, bins=20)
plt.title("Histogram - Distribution of Total Bill")
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pair Plot - All Numeric Relationships", y=1.02)
plt.show()
```
