# Different visualizations on Matplotlib

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Prime Numbers")
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.barh(categories, values, color='green')
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['red', 'yellow', 'blue', 'green']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.axis('equal')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='orange')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.figure(figsize=(8, 5))
plt.boxplot(data, vert=True, patch_artist=True)
plt.title("Box Plot Example")
plt.xlabel("Data Group")
plt.ylabel("Values")
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
product_a = [3, 4, 6, 7, 8]
product_b = [1, 2, 4, 5, 6]

plt.figure(figsize=(8, 5))
plt.stackplot(days, product_a, product_b, labels=['Product A', 'Product B'], colors=['#FF9999', '#66B2FF'])
plt.title("Area Plot Example")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend(loc='upper left')
plt.show()
```

```python
matrix_data = np.random.rand(5, 5)

plt.figure(figsize=(6, 6))
plt.imshow(matrix_data, cmap='coolwarm', interpolation='nearest')
plt.title("Heatmap Example")
plt.colorbar()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine")

axs[0, 1].plot(x, np.cos(x), color='r')
axs[0, 1].set_title("Cosine")

axs[1, 0].scatter(np.random.rand(50), np.random.rand(50))
axs[1, 0].set_title("Scatter")

axs[1, 1].bar(['A', 'B', 'C'], [5, 7, 3])
axs[1, 1].set_title("Bar")

fig.suptitle("Multiple Subplots Example")
plt.tight_layout()
plt.show()
```
